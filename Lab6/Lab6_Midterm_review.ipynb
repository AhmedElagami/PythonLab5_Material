## OOP vs Procedural

- **Procedural Programmin**g is a linear approach to programming. A program is divided into small parts called functions. The main focus is on functions rather than data.

- **Object-Oriented Programmin**g, on the other hand, bundles the data and the functions that operate on the data as a single unit called an object. This makes it easy to design more complex systems, as it modularizes the problem into simpler, manageable units.

# IO

Modes:
- 'r': Read (default). Opens the file for reading.

- 'w': Write. Opens the file for writing (creates a new file or truncates an existing file).

- 'x': Exclusive creation. Creates a new file and fails if the file already exists.

- 'a': Append. Opens the file for writing, appending to the end if the file exists.

- 'b': Binary. Opens the file in binary mode for reading or writing (e.g., 'rb' or 'wb').

- 't': Text. Opens the file in text mode (default, used with other modes).

- '+': Updating. Opens the file for both reading and writing.

## Error Codes

https://docs.python.org/3/library/errno.html

- **errno** is a standard module in Python that defines numbers for error codes produced by the operating system.

- The exception IOError, for example, has an **errno** attribute, which is a numeric error code corresponding to the error that occurred at the system level.

- This errno attribute matches one of the error codes defined in the errno module.

```python
import errno

try:
    # Attempt to open a file that doesn't exist
    with open('nonexistent_file.txt') as f:
        content = f.read()
except IOError as e:
    # Check the error code
    if e.errno == errno.ENOENT:
        print("Error: the file does not exist.")
    elif e.errno == errno.EACCES:
        print("Error: you don't have permission to read the file.")
    else:
        print("Some other error occurred:", e.errno)
```

## bytearray

```python
# Read an image into a bytearray
input_image_path = 'input_image.jpg'
output_image_path = 'output_image.jpg'

# Read the input image into a bytearray
with open(input_image_path, 'rb') as file:
    image_data = bytearray(file.read())

# do something

# Write the bytearray back out to a new image file
with open(output_image_path, 'wb') as file:
    file.write(image_data)

print(f"Image saved to {output_image_path}")
```

# Bitwise Operators

https://realpython.com/python-bitwise-operators/#bitwise-xor

```python
x = 1234 # a decimal number - a system to the power of 10
# 1000 + 200 + 30 + 4
# 1 * 10^3   +  2 * 10^2  + 3 * 10^1  +   4 * 10^0 = 1234
print(bin(x))

y = 0b10011010010 # a binary number - a system to the power of 2
#   (1 * 2^10) + (0 * 2^9) + (0 * 2^8) + (1 * 2^7) + (1 * 2^6) + (0 * 2^5)
# + (1 * 2^4) + (0 * 2^3) + (0 * 2^2) + (1 * 2^1) + (0 * 2^0) =  1234

print(x == y)
print(int(y))
```

```python
x = 4  # Binary representation: 0 1 0 0
y = 1  # Binary representation: 0 0 0 1

# AND: if both of them are 1, then we get 1
a = x & y  # Result: 0b0000

# OR: if one of them is 1, then we get 1
b = x | y  # Result: 0b0101

# NOT: in python you get this value ()
# Positive Numbers: it seems to return the negative of the number after increasing by 1.
# Negative Numbers: it returns the positive number after decreasing by 1.
# Most Significant Bit : This is the leftmost bit
# Least Significant Bit: This is the rightmost bit
c = ~x  # Result: 0b11111111111111111111111111111011 (two's complement of -5)

# XOR: if one of them is 1, BUT NOT BOTH, then we get 1
# 5 is            0   1    2   5        =>  0 1 0 1
#                                           0 1 0 0
d = x ^ 5  # Result: 0b0001

# shift to the right by 2 bits, filled with zeros
e = x >> 2  # Result: 0b0001

# shift to the left by 2 bits, filled with zeros
f = x << 2  # Result: 0b10000

# Print the results of the bitwise operations
print(a, b, c, d, e, f)  # Expected output: 0 5 -5 (result of x^5) 1 16
```

## map/ filter



```python
numbers = [1, 2, 3, 4]

# Using map with a predefined function
def square(num):
    return num * num

squared_numbers = map(square, numbers)

# Using map with a lambda function
squared_numbers_lambda = map(lambda num: num * num, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16]
print(list(squared_numbers_lambda))  # Output: [1, 4, 9, 16]
```

```python
numbers = [1, 2, 3, 4, 5, 6]

# Using filter with a predefined function
def is_even(num):
    return num % 2 == 0

even_numbers = filter(is_even, numbers)

# Using filter with a lambda function
even_numbers_lambda = filter(lambda num: num % 2 == 0, numbers)

print(list(even_numbers))  # Output: [2, 4, 6]
print(list(even_numbers_lambda))  # Output: [2, 4, 6]
```
