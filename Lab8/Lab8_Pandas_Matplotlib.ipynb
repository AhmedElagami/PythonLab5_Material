### Basic Graph

## How to style plots?

https://www.tutorialspoint.com/numpy/numpy_matplotlib.htm

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(8,5), dpi=100)

# Line 1
ax.plot(x, y, 'b^--', label='2x')

# Line 2
x2 = np.arange(0,4.5,0.5)
ax.plot(x2[:6], x2[:6]**2, 'r', label='X^2') # Plot part of the graph as line
ax.plot(x2[5:], x2[5:]**2, 'r--') # Plot remainder of graph as a dot

# Adding a title and setting font
ax.set_title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# Setting X and Y labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Setting X and Y axis tick marks
ax.set_xticks([0,1,2,3,4])
# ax.set_yticks([0,2,4,6,8,10])

# Adding a legend
ax.legend()

# Saving the figure
fig.savefig('mygraph_oo.png', dpi=300)

# Display the plot
plt.show()
```

### Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(8, 5), dpi=100)

# Line 1
ax.plot(x, y, 'b^--', label='2x')

# Line 2
x2 = np.arange(0, 4.5, 0.5)
# Plot part of the graph as line
ax.plot(x2[:6], x2[:6]**2, 'r', label='X^2')
# Plot remainder of graph as a dot
ax.plot(x2[5:], x2[5:]**2, 'r--')

# Adding a title and setting font
ax.set_title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# Setting X and Y labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Setting X and Y axis tick marks
ax.set_xticks([0, 1, 2, 3, 4])
# ax.set_yticks([0, 2, 4, 6, 8, 10])

# Adding a legend
ax.legend()

# Show plot
plt.show()

# Save figure (dpi 300 is good when saving so graph has high resolution)
fig.savefig('mygraph_oo.png', dpi=300)
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
values = [1, 4, 2]

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(6, 3), dpi=100)

# Create bars using the axis object
bars = ax.bar(labels, values)

# Set patterns for each bar
patterns = ['/', 'O', '*']
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)

# Save the figure
fig.savefig('barchart_oo.png', dpi=300)

# Show the plot
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(6,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

### Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

### Pie Chart

## Show Foot(Left, Right) Preference of FIFA Players

image.png

#### Pie Chart #1

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

## Show Weight of Professional Soccer Players

image.png

#### Pie Chart #2

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```
